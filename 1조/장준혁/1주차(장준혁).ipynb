# 간단한 분석

* pandas의 pivottable 기능을 이용해서 각자 자료를 3개씩 분석해 주세요

* 예시 
* 분석한 내용
> * tp09열은 Calm, emotionally stable을 나타내는데 내 생각에 남성이 여성보다 더 stable 할 것 같다
> * 따라서 성별로 tp09에 어떻게 답하였는지 사이즈를 살펴보았다. (0에 가까울 수록 강력하게 동의)

```python
pivot = pd.pivot_table(data = data, index = "tp09", columns = "gender", aggfunc = "size", fill_value = 0)
```

```python
pivot
# 결과를 보니 강력하게 동의하는 0~1에 남성들이 많았다.
```

* 이 이후에는 0,1로 강력하게 답한 사람, 중간으로 답한 사람, 6,7로 약하게 답한 사람별로 나눠서 추가적으로 분석해 볼 수도 있겠네요
* 이런 식으로 꼬리를 물어가면서 분석하면 조금 더 다양한 결과가 나올 것 같아요.

## 첫번째 분석

* 분석한 내용 간단히 설명 : 학력의 경우 오름차순으로 교육수준이 높음(0 제외)이며 QqA항목은 대부분의 사람들이 대체로 친절하고 착하다를 나타내는 항목(5에 가까울수록 강력한 동의). 분석 결과 교육수준이 높아질수록 1.0에 대비 5.0 비율이 높게 나오는 것을 확인할 수 있음. 교육수준이 높을수록 대다수의 사람들이 착하다고 믿는 비율이 높아짐

```python
pivot_1 = pd.pivot_table(data=data, index="education", columns="QqA", aggfunc= "size", fill_value=0)
```

```python
pivot_1
```

## 두번째 분석

* 분석한 내용 간단히 설명 : QbA 항목은 범죄자들이 검거되는 이유는 범죄자들이 일반인들에 비해 stupid 하다는 항목(5에 가까울수록 강력한 동의). 고졸 이하의 교육수준에서는 Disagree보다 Agree비율이 더 높음. 반대로 대졸 이상의 학력에서는 Disagree가 압도적으로 높은 결과를 보여줌

```python
pivot_2 = pd.pivot_table(data=data, index="education", columns="QbA", aggfunc= "size", fill_value=0)
```

```python
pivot_2
```

## 세번째 분석

* 분석한 내용 간단히 설명 : QfA 항목은 다른 사람을 속이는 데에는 변명의 여지가 없다는 항목. 비종교인에 비해 종교인들이 Disagree 대비 Agree 비율이 훨씬 높다는 걸 발견할 수 있음. 특히나 기독교 계열의 신자의 경우 타 종교 대비 해당 비율이 더 높은 것으로 보임

```python
pivot_3 = pd.pivot_table(data=data, index="religion", columns="QfA", aggfunc= "size", fill_value=0)
```
