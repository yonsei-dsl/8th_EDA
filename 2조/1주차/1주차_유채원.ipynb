```python
import numpy as np
import pandas as pd
```

```python
appdata = pd.read_csv('googleplaystore.csv')
review = pd.read_csv('googleplaystore_user_reviews.csv')
```

```python
appdata
```

```python
appdata = appdata.replace('Varies with device', np.nan).dropna()
```

```python
appdata = appdata.drop('Type', axis=1)
```

```python
appdata.iloc[:,:10]
```

```python
appdata['Price'].unique()
```

```python
appdata['Installs'] = np.array(x[:-1] for x in list(appdata['Installs']))
```

```python
def size_converter(x):
    if 'M' in x:
        return(float(x[:-1])*1000000)
    elif 'k' in x:
        return(float(x[:-1])*1000)

def price_converter(x):
    if x != '0':
        return (float(x[1:]))
    else:
        return(0)
def install_converter(x):
    return float(x.replace(',',''))
```

```python
appdata['Size']= appdata['Size'].apply(size_converter)
appdata['Price'] = appdata['Price'].apply(price_converter)
appdata['Installs'] = appdata['Installs'].apply(install_converter)
```
