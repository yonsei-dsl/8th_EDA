```python
import numpy as np
import pandas as pd
import csv
```

```python
data = pd.read_csv('googleplaystore.csv', names = ['App','Category','Rating','Reviews','Size','Installs','Type','Price','Content','Genres','Last Update','Current Version','Android Version'],encoding = 'UTF-8')
```

```python
datarv= pd.read_csv('googleplaystore_user_reviews.csv',names = ['App','Translated','Sentiment','Sentiment_Polarity','Sentiment_Subjectivity'],encoding = 'UTF-8')
```

```python
mgdata = pd.merge(data,datarv,how = 'outer')
```

```python
mgdata.App
```

```python
mgdata.iloc[:, [2,3]]
```

```python
print(mgdata.isnull().sum(axis=0)) 
print('-'*30)

print(mgdata.isnull().sum(axis=1)) 
```

```python
mgdata.dropna(subset = ['Translated'])
mgdata['Rating'].interpolate(method = 'linear')
mgdata.dropna(subset = ['Rating'])
```

```python
mgdata['Rating']=mgdata['Rating'].astype(str)
mgdata.groupby(['App','Installs']).agg({'Rating':['sum','max','min']})
```

```python
pd.pivot_table(mgdata,index = 'App',values = ['Rating','Installs'],aggfunc = {'Rating':['sum','min','max'],'Installs':'count'})
```
