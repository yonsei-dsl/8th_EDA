### 분석내용

- 4개 지역 모두에서 소매업, 음식업, 생활서비스업이 가장 많은 비중을 차지
- 광역시(광주 & 대구)는 음식업과 소매업이 비슷한 수준
- 전북, 전남 지역은 음식업보다 소매업이 월등히 높음
- (광주와 대구 등 광역시에서는 전북, 전남 지역에 비해 음식업에 종사하는 소상공인들이 많아 보인다.)
- 대체적으로 숙박업, 스포츠업에 종사하는 소상공인들이 적음

1. 전북 & 전남:
- 인구 / 매장수: 전남 > 전북 (큰 차이 없음)
- 1인당 매장수: 큰 차이없음
- 소매업 > 음식업 > 생활서비스업
- 지역(시군구)마다 매장수 편차가 큼
- 전북 최하위 업종: 숙박, 스포츠
- 전남 최하위 업종: 부동산, 스포츠 (전남은 다른 지역에 비해 숙박업에 종사하는 사람들이 많아 보인다.)

2. 광주 & 대구
- 인구 / 매장수 : 대구 > 광주
- 1인당 매장수: 광주 > 대구
- 광주: 소매업 > 음식업 > 생활서비스업
- 대구: 음식업 > 소매업 > 생활서비스업
- 두 광역시 모두 숙박업과 스포츠업이 최하위 업종
